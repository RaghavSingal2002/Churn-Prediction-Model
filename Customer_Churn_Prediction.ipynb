# Data Cleaning 

1. Create a copy of base data for manupulation & processing 

```python
telco_data = telco_base_data.copy()
```

2.Total Charges should be numberic amount.Let's convert it to numeric data type

```python
telco_data['TotalCharges'] = pd.to_numeric(telco_data['TotalCharges'], errors='coerce')
telco_data.isnull().sum()
```

3.As we can see that there are 11 missing values in TotalCharges column.Let's check these records.

```python
telco_data.loc[telco_data['TotalCharges'].isnull() == True]
```

4.Missing value Treatment-
Since the % of these records compared to toal dataset is very low ie 0.15%,it is safe to ignore them from further processing.

```python
# Removing the missing values
telco_data.dropna(how = 'any',inplace = True)
```

5.Divide Customers into bins based on tenure eg.for tenure  < 12 months:assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24;so on...

```python
# Get the max tenure 
print(telco_data['tenure'].max())
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]
telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
telco_data['tenure_group'].value_counts()
```

6.Remove columns not required for processing

```python
telco_data.drop(columns = ['customerID', 'tenure'], axis = 1, inplace = True)
telco_data.head()
```

# Data Exploration

1.Plot the distribution of individual predictors by churn 

# Univariate Analysis

```python
for i, predictor in enumerate(telco_data.drop(columns=['Churn','TotalCharges','MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=telco_data, x=predictor, hue='Churn')
```

2.Convert the target variable 'Churn' in a binary numeric variable ie Yes=1; No = 0

```python
telco_data['Churn'] = np.where(telco_data.Churn == 'Yes',1,0)
```

```python
telco_data.head()
```

3.Convert all the categorical variables into numerical variables.

```python
telco_data_dummies = pd.get_dummies(telco_data)
telco_data_dummies.head()
```

Relationship between Monthly Charges and Total Charges

```python
sns.lmplot(data=telco_data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

Total Charges increases as Monthly Charges increases as expected.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Plot KDE for churn=0 (No Churn)
ax1 = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies['Churn'] == 0)], color="Red", fill=True)
# Plot KDE for churn=1 (Churn)
ax2 = sns.kdeplot(telco_data_dummies.MonthlyCharges[(telco_data_dummies['Churn'] == 1)], ax=ax1, color="Blue", fill=True)

# Set labels and title for the plot
plt.legend(["No Churn", "Churn"], loc='upper right')
plt.xlabel('Monthly Charges')
plt.ylabel('Density')
plt.title('Monthly Charges by Churn')
```

# Bivariate Analysis

```python
new_df1_target0 = telco_data.loc[telco_data["Churn"] == 0]
new_df1_target1 = telco_data.loc[telco_data["Churn"] == 1]
```

```python
def uniplot(df, col, title, hue= None):
    
    sns.set_style('Whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width, 8)
    plt.xticks(rotation = 45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x = col, order = df[col].value_counts().index, hue = hue, palette = 'bright')
    
    plt.show()
```

```python
sns.catplot(data=new_df1_target0, x='Partner', kind='count', hue='gender', col='Partner', palette='viridis', aspect=0.8)
plt.title('Distribution of Gender for Non Churned Customers')
plt.show()
```

```python
sns.catplot(data=new_df1_target1, x='Contract', kind='count', hue='gender', palette='viridis')
plt.title('Distribution of Contract for Churned Customers')
plt.show()
```

```python
sns.catplot(data=new_df1_target1, x='TechSupport', kind='count', hue='gender', palette='viridis')
plt.title('Distribution of TechSupport for Churned Customers')
plt.show()
```

```python
sns.catplot(data=new_df1_target1, x='SeniorCitizen', kind='count', hue='gender', palette='viridis')
plt.title('Distribution of SeniorCitizen for Churned Customers')
plt.show()
```
